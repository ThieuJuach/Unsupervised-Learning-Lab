```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
```

```python
# Load the dataset

df = pd.read_csv('Mall_Customers.csv')
```

```python
# Display basic information
df.info()
```

```python
# Display the first few rows
df.head()
```

# Data Visualization and Preprocessing

```python
# Basic statistics of the dataset
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Dropping any missing values
df.dropna(inplace=True)
```

```python
# Normalization/Scaling (StandardScaler or MinMaxScaler)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
# Convert the scaled features back into a DataFrame for visualization and exploration
scaled_df = pd.DataFrame(scaled_features, columns=['Annual Income', 'Spending Score'])
```

```python
# Pair plots to visualize relationships
sns.pairplot(scaled_df)
plt.show()
```

```python
# Histograms to view the distribution of the data
scaled_df.hist(figsize=(10, 5), bins=30)
plt.show()
```

# K-Means Clustering

```python
# Elbow Method to determine optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# KMeans with optimal number of clusters (let's assume k=5 from the elbow method)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_df)
```

```python
# 2D Scatter plot using PCA (to visualize clusters)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_df)
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title('K-Means Clustering (PCA Reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

# Hierarchical Clustering


```python
# Hierarchical Clustering using Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=5, linkage='ward')
agg_labels = agg_clustering.fit_predict(scaled_df)
```

```python
# Dendrogram
Z = linkage(scaled_df, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
# Compare linkage methods ('ward', 'average', 'complete')
for linkage_method in ['ward', 'average', 'complete']:
    agg_clustering = AgglomerativeClustering(n_clusters=5, linkage=linkage_method)
    agg_labels = agg_clustering.fit_predict(scaled_df)
    plt.scatter(pca_data[:, 0], pca_data[:, 1], c=agg_labels, cmap='rainbow')
    plt.title(f'Agglomerative Clustering - {linkage_method}')
    plt.show()
```

# Dimensionality Reduction (PCA)


```python
# PCA for dimensionality reduction
pca = PCA()
pca.fit(scaled_df)
explained_variance = pca.explained_variance_ratio_
```

```python
# Plot explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Explained Variance Ratio by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Ratio')
plt.show()
```

```python
# Visualize data on first two principal components
pca_2d = PCA(n_components=2).fit_transform(scaled_df)
plt.figure(figsize=(8, 6))
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title('PCA with K-Means Clusters')
plt.show()
```

# Dimensionality Reduction (t-SNE)


```python
# t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
tsne_data = tsne.fit_transform(scaled_df)
```

```python
# Plot t-SNE result
plt.figure(figsize=(8, 6))
plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title('t-SNE with K-Means Clusters')
plt.show()
```

# Silhouette Score

```python
# Calculate Silhouette Score for K-Means
kmeans_silhouette = silhouette_score(scaled_df, kmeans_labels)
print(f'Silhouette Score for K-Means: {kmeans_silhouette}')
```

```python
# Silhouette Score for Hierarchical Clustering
agg_silhouette = silhouette_score(scaled_df, agg_labels)
print(f'Silhouette Score for Hierarchical Clustering: {agg_silhouette}')
```
